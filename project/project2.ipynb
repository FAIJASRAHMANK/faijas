## Feature as age

```python
# Plot age distribution with histogram and KDE
sns.displot(data=s, x='age', bins=12, kde=True, color='mediumseagreen')

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python

bins = [18, 30, 40, 50, 60, 70, 100]
labels = ['18â€“30', '31â€“40', '41â€“50', '51â€“60', '61â€“70', '71+']
s['age_group'] = pd.cut(s['age'], bins=bins, labels=labels)

# Plot count of each age group
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='age_group', data=s, order=labels,palette='Set2')

# Add count labels
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])
ax.bar_label(ax.containers[3])
ax.bar_label(ax.containers[4])
ax.bar_label(ax.containers[5])


plt.title('Count of Customers in Each Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
max_age = s['age'].max()
print("Maximum age is:", max_age)
```

```python
# Step 3: Plot grouped counts (y = 0 vs 1) by age group
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=s, x='age_group', hue='y', order=labels, palette='Set2')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title('Subscription Counts by Age Group ')
plt.xlabel('Age Group')
plt.ylabel('Customer Count')
plt.legend(title='Subscribed (y)')
plt.tight_layout()
plt.show()
```

```python
#The analysis of the age feature reveals that most customers fall within the 30 to 60 age range, with the highest concentration between 31â€“50 years.
#Subscription rates to term deposits vary significantly by age group,
# with middle-aged customersâ€”particularly those aged 41â€“60â€”showing a higher likelihood of subscribing compared to younger (18â€“30) or older (71+) individuals. 
#This indicates that age is an important predictor in determining customer behavior.
#The bank can use this insight to better target its marketing efforts toward age groups with higher subscription probabilities,
#thereby improving campaign efficiency
```

```python
# Remove outliers from 'balance'


# Step 1: Copy s to s_clean
s_clean = s.copy()

# Step 2: Remove outliers from 'balance'
Q1_bal = s['balance'].quantile(0.25)
Q3_bal = s['balance'].quantile(0.75)
IQR_bal = Q3_bal - Q1_bal
s_clean = s_clean[(s_clean['balance'] >= Q1_bal - 1.5 * IQR_bal) & 
                  (s_clean['balance'] <= Q3_bal + 1.5 * IQR_bal)]

# Step 3: Remove outliers from 'campaign'
Q1_cam = s['campaign'].quantile(0.25)
Q3_cam = s['campaign'].quantile(0.75)
IQR_cam = Q3_cam - Q1_cam
s_clean = s_clean[(s_clean['campaign'] >= Q1_cam - 1.5 * IQR_cam) & 
                  (s_clean['campaign'] <= Q3_cam + 1.5 * IQR_cam)]

# Step 4: Remove outliers from 'age'
Q1_age = s['age'].quantile(0.25)
Q3_age = s['age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
s_clean = s_clean[(s_clean['age'] >= Q1_age - 1.5 * IQR_age) & 
                  (s_clean['age'] <= Q3_age + 1.5 * IQR_age)]

# Step 5: Remove outliers from 'previous'
Q1_pre = s['previous'].quantile(0.25)
Q3_pre = s['previous'].quantile(0.75)
IQR_pre = Q3_pre - Q1_pre
s_clean = s_clean[(s_clean['previous'] >= Q1_pre - 1.5 * IQR_pre) & 
                  (s_clean['previous'] <= Q3_pre + 1.5 * IQR_pre)]

```
